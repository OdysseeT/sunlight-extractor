```python
import requests
import xmltodict
import pandas as pd
import numpy as np
import datetime
import simplekml
```

```python
URL = "http://www.solartopo.com/services/getDetailData.php"
```

```python
df = pd.DataFrame (columns=['latitude','longitude','date','altitude','sunrise','sunset','sunset_lat','sunset_long','sunrise_lat','sunrise_long'])
```

```python
lat_start = 46.38
step = 0.005
lat_end = lat_start-(83*step)

long_start = 5.93
long_end = (140*step)+long_start

date = "5-5-2020"

print("Latitude: from {} to {}".format(lat_start, lat_end))
print("Longitude: from {} to {}".format(long_start, long_end))
```

```python
def request_data(lat, long, date):
    r = requests.post(URL, data = {'lat':lat, 'long':long,'date':date})

    doc = xmltodict.parse(r.text)
    return {
        'latitude':lat,
        'longitude':long,
        'date':date,
        'altitude':doc['Response']['altitude'],
        'sunrise':doc['Response']['sunrise'],
        'sunset':doc['Response']['sunset'],
        'sunset_lat':doc['Response']['sunset_lat'],
        'sunset_long':doc['Response']['sunset_long'],
        'sunrise_lat':doc['Response']['sunrise_lat'],
        'sunrise_long':doc['Response']['sunrise_long']}
```

```python
for month in range(1,12,1):
    date = "21-{}-2020".format(month)
    
    for lat in np.arange(lat_start, lat_end, step*-1):
        lat = round(lat, 5)
        print("Request for lat: {} - Date: {}".format(lat,date))
        
        for long in np.arange(long_start, long_end, step):
            long = round(long, 5)

            data = request_data(lat, long, date)
            df = df.append(data, ignore_index=True)
        
```

```python
df.head()
```

```python
df.sunset = df.sunset.astype('float64')
df.sunrise = df.sunrise.astype('float64')
df.sunset_lat = df.sunset_lat.astype('float64')
df.sunset_long = df.sunset_long.astype('float64')
df.sunrise_lat = df.sunrise_lat.astype('float64')
df.sunrise_long = df.sunrise_long.astype('float64')
df.altitude = df.altitude.astype('int16')
```

```python
df["daylength"] = df.sunset-df.sunrise
```

```python
df["daylength_hours"] = df["daylength"].apply(lambda x: datetime.timedelta(seconds=round(x)))
```

```python
df.to_csv("data/daylight.csv", index=False)
```
